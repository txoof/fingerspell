```python
import pandas as pd
from pathlib import Path
```

```python
my_data = pd.read_csv('./data/dataset/ngt_static_landmarks.csv')
aaron_data = pd.read_csv('./data/ngt_static_landmarks.csv')
```

```python
combined = pd.concat([my_data, aaron_data], ignore_index=True)
```

```python
print(f"My data: {len(my_data)} samples")
print(f"Teammate data: {len(aaron_data)} samples")
print(f"Combined: {len(combined)} samples")
print(f"\nSamples per letter:")
print(combined['letter'].value_counts().sort_index())
```

```python
combined.to_csv('./data/dataset/ngt_static_landmarks_1807.csv', index=False)
```
