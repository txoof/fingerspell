```python
import csv

# First, check what's in the original file
original_count = 0
label_counts = {}
with open('../static_collection/ngt_keypoint.csv', 'r', encoding='utf-8-sig') as f:
    for row in csv.reader(f):
        if row:
            original_count += 1
            label = int(row[0])
            label_counts[label] = label_counts.get(label, 0) + 1

print(f"Original file has: {original_count} samples")
print(f"Dynamic letters count:")
print(f"  H(7): {label_counts.get(7, 0)}")
print(f"  J(9): {label_counts.get(9, 0)}")
print(f"  U(20): {label_counts.get(20, 0)}")
print(f"  X(23): {label_counts.get(23, 0)}")
print(f"  Z(25): {label_counts.get(25, 0)}")

# Indices of dynamic letters to remove: H(7), J(9), U(20), X(23), Z(25)
dynamic_indices = {7, 9, 20, 23, 25}

# Read and filter data
cleaned_data = []
removed_count = 0
with open('../static_collection/ngt_keypoint.csv', 'r', encoding='utf-8-sig') as f:
    for row in csv.reader(f):
        if row:  # Skip empty rows
            label = int(row[0])
            if label not in dynamic_indices:
                cleaned_data.append(row)
            else:
                removed_count += 1

# Write cleaned dataset
with open('../static_collection/ngt_keypoint_static.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerows(cleaned_data)

print(f"\nCleaned samples: {len(cleaned_data)}")
print(f"Removed: {removed_count} samples")
print(f"Total check: {len(cleaned_data)} + {removed_count} = {len(cleaned_data) + removed_count}")
```
